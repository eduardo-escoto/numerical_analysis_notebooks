### Problem 1

#### Part a)
Equaiting the leading coefficient of the Lagrange form of the interpolation polynomial $p_{n}(x)$ with that of the Newton's form deduce that

\begin{equation}
f[x_{0}, x_{1},\dots,x_{n}] = \sum_{j=0}^{n}\frac{f(x_{j})}{\displaystyle{\prod_{k=0,k\neq j}^{n}}(x_{j}-x_{k})}
\end{equation}

##### Solution
Notice that by definition:
$$f[x_{0}, x_{1},\dots,x_{n}] = \frac{p_{n}(x) - p_{n-1}(x)}{(x-x_{0})\dots(x-x_{n-1})}$$
Additionally, from Lagrange Form of the integrating polynomial, we have:
$$p_{n} = \sum^{n}_{j}l_{j}(x)f_{j}$$
Thus we have:
$$f[x_{0}, x_{1},\dots,x_{n}] = \sum^{n}_{j}l_{j}(x)f_{j} - \sum^{n-1}_{j}l_{j}(x)f_{j} \implies f[x_{0}, x_{1},\dots,x_{n}] =   \frac{l_{n}(x)f_{n}}{(x-x_{0})\dots(x-x_{n-1})}$$
Now, notice that for $f_{n} = f(x_{n})$ we have by the interpolation property:
$$f(x_{n}) = p_{n}(x_{n}) = f[x_{0}] + \dots f[x_{0},\dots,x_{n}](x-x_{0})\dots(x-x_{n-1}) $$
Thus we have:
$$ f[x_{0}, x_{1},\dots,x_{n}] =   \frac{l_{n}(x)\left(f[x_{0}] + \dots f[x_{0},\dots,x_{n}](x-x_{0})\dots(x-x_{n-1})\right)}{(x-x_{0})\dots(x-x_{n-1})}$$
Which becomes:
$$f[x_{0}, x_{1},\dots,x_{n}] =  \frac{(x-x_{0})\dots(x-x_{n-1})}{{(x_{n}-x_{0})\dots(x_{n}-x_{n-1})}{(x-x_{0})\dots(x-x_{n-1})}}\left(f[x_{0}] + \dots f[x_{0},\dots,x_{n}](x-x_{0})\dots(x-x_{n-1})\right)$$
Then, by simplification, we see that the ${(x-x_{0})\dots(x-x_{n-1})}$ product cancels, leaving:
$$f[x_{0}, x_{1},\dots,x_{n}] =  \frac{1}{{(x_{n}-x_{0})\dots(x_{n}-x_{n-1})}}\left(f[x_{0}] + \dots f[x_{0},\dots,x_{n}](x-x_{0})\dots(x-x_{n-1})\right)$$
Then:
$$f[x_{0}, x_{1},\dots,x_{n}] =  \frac{f[x_{0}]}{{(x_{n}-x_{0})\dots(x_{n}-x_{n-1})}} + \dots \frac{f[x_{0},\dots,x_{n}](x-x_{0})\dots(x-x_{n-1})}{{(x_{n}-x_{0})\dots(x_{n}-x_{n-1})}}$$
Now, by the fact that:
$$f[x_{0}] = f_{0}\text{, and } f[x_{0},x_{1}] = \frac{f_1-f_0}{x_1-x_0}, \text{ and so on...}$$
Then the above equation will clearly simplify to:
$$f[x_{0}, x_{1},\dots,x_{n}] =  \frac{f_{0}}{{(x_{0}-x_{1})\dots(x_{0}-x_{n})}} + \dots + \frac{f_{n}}{{(x_{n}-x_{0})\dots(x_{n}-x_{n-1})}}$$
And, finally:
$$f[x_{0}, x_{1},\dots,x_{n}] = \sum_{j=0}^{n}\frac{f(x_{j})}{\displaystyle{\prod_{k=0,k\neq j}^{n}}(x_{j}-x_{k})}$$
Which is the result we wanted.

#### Part b)
Use $(1)$ to conclude that divided differences are symmetric functions of their arguments. i.e. any permutation of $x_{0},x_{1},\dots,x_{n}$ leaves the corresponding divided difference unchanged.

##### Solution
Let $x_{0},\dots,x_{n}$ be interpolation points. Thus, by above we have the divided difference to be:
$$f[x_{0}, x_{1},\dots,x_{n}] = \sum_{j=0}^{n}\frac{f(x_{j})}{\displaystyle{\prod_{k=0,k\neq j}^{n}}(x_{j}-x_{k})}$$
Now, if we let $y_{0},\dots,y_{n}$ be a permutation of the interpolations points, such that $y_{i}$ corresponds to some $x_{j}$ for $i,j = 0,\dots,n$ and $f(y_{i}) = f(x_{j})$. Where we denote the $x$ corresponding to $y_{j}$ as $x^{(y_{j})}$. We have:
$$f[y_{0}, y_{1},\dots,y_{n}] = \sum_{j=0}^{n}\frac{f(y_{j})}{\displaystyle{\prod_{k=0,k\neq j}^{n}}(y_{j}-y_{k})}$$
Which then becomes:
$$f[y_{0}, y_{1},\dots,y_{n}] =  \frac{f(y_{0})}{{(y_{0}-y_{1})\dots(y_{0}-y_{n})}} + \dots + \frac{f(y_{n})}{{(y_{n}-y_{0})\dots(y_{n}-y_{n-1})}}$$
Now, since each $y_{i}$ corresponds to some $x_{j}$ we have:
$$f[y_{0}, y_{1},\dots,y_{n}] =  \frac{f(x^{(y_{0})})}{{(x^{(y_{0})}-x^{(y_{1})})\dots(x^{(y_{0})}-x^{(y_{n})})}} + \dots + \frac{f(x^{(y_{n})})}{{(x^{(y_{n})}-x^{(y_{0})})\dots(x^{(y_{n})}-x^{(y_{n-1})})}}$$
Which then gives:
$$f[y_{0}, y_{1},\dots,y_{n}] = \sum_{j=0}^{n}\frac{f(x_{j})}{\displaystyle{\prod_{k=0,k\neq j}^{n}}(x_{j}-x_{k})}$$
Since it is merely just the sum over all the corresponding $x$ values. Thus, the order of the $x_{i}$ does not affect the divided difference.

### Problem 2
In Newton's form of the interpolation poluynomial we need to compute the coefficients $c_{0} = f[x_0], c_1 = f[x_0,x_1], \dots, c_n = f[x_0,x+1,\dots, x_n]$. In the table of divided differences we proceed column by column and the needed coefficients are in the uppermost diagonal. A simple 1D array, c, of size $n+1$, can be used to store and compute these values. We just have to compute theme from bottom to top to avoid losing values we have already computed. The following pseudocode does precisely this:<br/>
for $j = 0,1,\dots,n$ <br/>
&emsp; $c_j = f_j$ <br/>
end
for $k = 1,\dots,n$ <br/>
&emsp; for $j = n, n-1, \dots, k$ <br/>
&emsp; &emsp; $c_j = (c_j-c_{j-1})/(x_j-x_{j-k})$ <br/>
&emsp; end <br/>
end

The evaluation of the interpolation polynomial in Newton's form can then be done with the Horner-like scheme seen in class:

$p=c_n$ <br/>
for $j=n-1,n-2,\dots,0$ <br/>
&emsp; $p = c_j + (x-x_{j})\cdot p$ <br/>
end

#### Part a)
Write a computer code to compute the coefficients $c_0,c_1,\dots,c_n$ and to evaluate the corresponding interpolation polynomial at an arbitrary point $x$. Test your codes and turn in a run of your test.

##### Solution

```python
def NewtonFormCoeffs(f, X):
    n = len(X)
    c = np.array([f(x) for x in X])
    for k in range(1, n):
        for j in reversed(range(k, n)):
            c[j] = (c[j]-c[j-1])/(X[j]-X[j-k])
    return c
    
def NewtonFormPoly(c,X):
    n = len(X)
    
    def poly(x):
        p = c[-1]
        for j in reversed(range(n)):
            p = c[j] + (x-X[j])*p
        return p

    return poly
```

Below we test out our functions by using the example in the course notes, $f(x) = 1 + x^{2}$. We should get $c_{2}, c_{1},c_{0} = 1$, and $c_{j} = 0$ for $j > 2$.

```python
f = lambda x: 1+x**2
X = np.arange(0,10,1)
c = NewtonFormCoeffs(f, X)
c
```

Now we test it out against the candidate function and plot it.

```python
p = np.vectorize(NewtonFormPoly(c,X))
points = np.linspace(-100, 100, 1000)

plt.figure(figsize=(9,6))
plt.plot(points, p(points), linewidth = 3,label = "$p_{n}(x)$")
plt.plot(points, f(points),linewidth = 3, linestyle = "--",alpha = .8, label = "f(x)")
plt.title("$x^{2} + 1$ vs Newton's Polynomial Form")
plt.legend();
```

As seen above, we end up getting the exact same function from our interpolation.

#### Part b)
Consider the function $f(x) = e^{-x^{2}}$ for $x\in[-1,1]$ and the nodes $x_j = -1+ j(2/10), j = 0,1,\dots,10$. Use your code in Part (a) to evaluate $p_{10}(x)$ at the points $\bar{x}_{j} = -1 + j(2/100), j = 0,1,\dots,100$ and plot the error $f(x)-p_{10}(x)$.

##### Solution

```python
f = lambda x: np.e**(-(x**2))
nodes = lambda n, start = -1 : np.array([start + j*(2/n) for j in range(n+1)])

X = nodes(10)
c = NewtonFormCoeffs(f, X)
p = NewtonFormPoly(c,X)
```

```python
P = np.vectorize(p)
points = nodes(100)

plt.figure(figsize=(9,6))
plt.plot(points, P(points), linewidth = 3,label = "$p_{10}(x)$")
plt.plot(points, f(points),linewidth = 3, linestyle = "--",alpha = .8, label = "f(x)")
plt.title("$e^{-x^{2}}$ vs Newton's Polynomial Form")
plt.legend();
```

```python
points = nodes(100)
err = np.vectorize(lambda x: f(x) - p(x))

plt.figure(figsize=(9,6))
plt.plot(points, err(points), linewidth = 3)
plt.title("$e^{-x^{2}}- p_{10}$");
```

Here we can see that the error is very, very small!

### Problem 3
*Inverse Interpolation*

##### Solution
Below we calculate the polynomials at $p_{1}$ and $p_{2}$ and also the $\bar{x}$ which we believe to be a better approximation to zero.

```python
f =lambda x: x - (np.e**(-x))

y_0 = -0.106530659712633
x_0 = 0.5

y_1 = 0.051188363905973 
x_1 = 0.6

inv_f_0_1 = (x_1-x_0)/(y_1-y_0)
inv_f_0_1

p_1 = lambda y: x_0 + (inv_f_0_1*(y - y_0))

x_2 = p_1(0)
y_2 = f(x_2)
inv_f_1_2 = (x_2-x_1)/(y_2-y_1)

inv_f_0_1_2 = (inv_f_1_2-inv_f_0_1)/(x_2-x_0)

p_2 = lambda y: x_0 + (inv_f_0_1*(y - y_0)) + (inv_f_0_1_2*(y-y_0)*(y-y_1))

bar_x = p_2(0)
```

So our better approximation for $f(x) = 0$ is $x = 0.5669079233508125$

Where we then plot our polynomial approximations in comparison to the original function.

```python
X = np.linspace(.567-.07,.567+.035,100)
P_1 = np.vectorize(p_1)
P_2 = np.vectorize(p_2)
F = np.vectorize(f)

plt.figure(figsize=(9,6))
plt.plot(X, F(X), label = "$f(x)$")
plt.hlines(y = 0, xmin = .567-.07, xmax=.567+.035, linestyle = "--" , 
           color = "red", label = "y = 0", alpha = .5)
plt.vlines(x = bar_x, ymin = min(F(X)), ymax=max(F(X)), linestyle = "--" , 
           color = "red", label = "x = ${}$".format(bar_x), alpha =.5)
plt.scatter(x_0, y_0, label = "$(x_{0},y_{0})$")
plt.scatter(x_1, y_1, label = "$(x_{1},y_{1})$")
plt.scatter(x_2, y_2, label = "$(x_{2},y_{2})$")
plt.scatter(bar_x, f(bar_x), label = "($\\bar{x} ,f\\bar{x})$")
plt.legend()
plt.title("$f(x)$ with inverse interpolation points");
```

### Problem 4
Obtain the Hermite Interpolation polynomial corresponding to the data $f(0) = 0$, $f'(0) = 0$, $f(1) = 2$, $f'(1) = 3$.

##### Solution
By using the Hermite interpolation table in equation $3.60$, we arrive at the interpolating polynomial:
$$p(x) = 3x^{2}-x^{3}$$
and then by differentiating $p$ we have:
$$p'(x) = 6x - 3x^{2}$$

Below we plot the interpolating polynomials with the interpolation data to show that interpolation property holds and it is the correct function!

```python
p = lambda x: 3*(x**2) - x**3
p_prime = lambda x:6*x - 3*(x**2)
X = np.linspace(-1.5,1.5,1000)
P = np.vectorize(p)
P_prime = np.vectorize(p_prime)

X_data = [0,1]
X_prime_data = [0,1]

Y_data = [0,2]
Y_prime_data = [0,3]

plt.figure(figsize=(9,6))
plt.plot(X, P(X), label = "$p(x)$")
plt.plot(X, P_prime(X), label = "$p'(x)$")
plt.vlines(x = 0, ymin = min([min(P(X)),min(P_prime(X))]), ymax = max([max(P(X)),max(P_prime(X))]),
          linestyle = "--" , color = "red", label = "y = 0", alpha = .5)
plt.hlines(y = 0, xmin = min(X), xmax = max(X),linestyle = "--" , 
           color = "red", label = "x = 0", alpha = .5)
plt.scatter(X_data,Y_data, label = "$f$ Data")
plt.scatter(X_prime_data,Y_prime_data, label = "$f'$ Data")
plt.legend()
plt.title("$p(x)$ and $p'(x)$ with interpolation data");
```
